```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

model_path = "../model_cache/intent_classifier_roberta/"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)
model.eval()

print('model loaded')

queries = [

    # 0. chit chat
    "hello, how are you?",
    "what's the weather like today?",
    "do you like music?",
    # 1. Rõ ràng (Clear Intent)
    "red sweater",
    "find outfits for this shirt",
    "do you have a black version",
    
    # 2. Cụt lủn, tối nghĩa (Ambiguous/Short)
    "white sneakers",
    "shoes",
    "wide leg jeans",
    
    # 3. Ngữ cảnh ẩn, phụ thuộc session (Implicit Context)
    "match this",
    "what about blue?",
    "does it go well with boots?",
    
    # 4. Truy vấn dài, nhiễu thông tin (Long-tail)
    "summer floral dress for a beach vacation in hawaii",
    "i have a black leather biker jacket, what should i wear inside for a date night?",
    
    # 5. Phủ định, so sánh (Negation)
    "something like the previous one but cheaper and no wool",
    "not this color, give me red ones",
    
    # 6. Đa ý định (Composite Intent)
    "find me a red hoodie and some matching sweatpants",
    "I want a blue shirt and pants that fit"
]

for q in queries:
    inputs = tokenizer(q, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    
    probs = F.softmax(outputs.logits, dim=-1)[0]
    
    print(f"Q: '{q}'")
    for i, label in model.config.id2label.items():
        print(f"  - {label}: {probs[i].item():.4f}")
    print("-" * 50)
```
